### 1.1. Missing Values

```python
## Here we will check the percentage of nan values present in each feature
## 1 -step make the list of features which has missing values
features_with_na=[features for features in df.columns if df[features].isnull().sum()>1]
```

```python
features_with_na
```

```python
for feature in features_with_na:
    print(feature, np.round(df[feature].isnull().mean(),4),' % missing values') #Example, if 2 out of 10 values in a column are 
#null,taking the mean of isnull() will give the percentage of null values in the column
```

Are there missing values, yes? Can we directly drop the row that has the missing values?
Is there a relationship between the missing values and the dependent feature,Sales Price?

```python
for feature in features_with_na:
    data= df.copy()

    # let's make a variable that indicates 1 if the observation was missing or zero otherwise
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    
    # let's calculate the mean SalePrice where the information is missing or present
    data.groupby(feature)['SalePrice'].median().plot.bar() 
# Applying groupby() function to group the data on the feature and then taking the median of the Sales Price
# More on groupby : https://www.geeksforgeeks.org/python-pandas-dataframe-groupby/
    plt.title(feature)
    plt.show()
```

From the above, Features like LotFrontage, missing values have a higher contribution to the SalesPrice. 
Hence, we cannot ignore them and we will take care of them in the feature Engineering section

### 1.3.Distribution of the Numerical Variables

Numerical features can be of 2 types: Discrete and Continuous
Discrete -> Takes a fixed number of values(all whole numbers).In our case, we are assuming upto 25 values that these can take
Continuous -> Can assume any value

```python
#Discrete Features
discrete_features= [feature for feature in numerical_features if len(df[feature].unique())<25 and 
feature not in year_feature+['Id']]
print(' Number of Discrete variables is {}'.format(len(discrete_features)))
```

```python
discrete_features
```

```python
df[discrete_features].head()
```

Exploratory Data Analysis is the analysis of the features with respect to the dependent variable. Hence, we now need to 
compare discrete features with the independent feature

```python
for feature in discrete_features:
    data=df.copy()
    df.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
#Continuous features
continuous_features=[features for features in numerical_features if features not in discrete_features+year_feature+['Id']]
print("Number of continuous features {}".format(len(continuous_features)))
```

```python
## Lets analyse the continuous values by creating histograms to understand the distribution

for feature in continuous_features:
    data=df.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

```python
# We will use the logarithmic transformation
for feature in continuous_features:
    data= df.copy()
    if 0 in data[feature].unique():#Log(0) is not undefined
        pass
    else:
        data[feature]=np.log(data[feature])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```

### 1.6.Outliers

```python
#Boxplot can help in visualizing outliers (only for numerical features)
for feature in continuous_features:
    data=df.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.title(feature)
        plt.show()
```
